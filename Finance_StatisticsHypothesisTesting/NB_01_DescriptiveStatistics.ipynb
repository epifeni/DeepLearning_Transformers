### Relative and Cumulative Frequencies with plt.hist()

```python
import numpy as np
import matplotlib.pyplot as plt
np.set_printoptions(precision=4, suppress= True)
```

```python
pop.size
```

```python
sample.size
```

```python
# Histogram - Relative Frequencies - Population equally weighted
(np.ones(len(pop)) / len(pop))
```

```python
plt.figure(figsize = (12, 8))
plt.hist(pop, bins = 75, weights = np.ones(len(pop)) / len(pop))
plt.title("Relative Frequencies - Population", fontsize = 20)
plt.xlabel("Stock Returns 2017 (in %)", fontsize = 15)
plt.ylabel("Relative Frequency", fontsize = 15)
plt.show()
```

```python
# Pitfall: Using density=True - this does not give equally weighted relative frequencies
plt.figure(figsize = (12, 8))
plt.hist(pop, bins = 50, density = True)
plt.title("Relative Frequencies - Population", fontsize = 20)
plt.xlabel("Stock Returns 2017 (in %)", fontsize = 15)
plt.ylabel("Relative Frequency", fontsize = 15)
plt.show()
```

```python
# Histogram - Cumulative Absolute Frequencies - Population
plt.figure(figsize = (12, 8))
plt.hist(pop, bins = 50, density = False, cumulative = True)
plt.title("Cumulative Absolute Frequencies - Population", fontsize = 20)
plt.xlabel("Stock Returns 2017 (in %)", fontsize = 15)
plt.ylabel("Cumulative Absolute Frequency", fontsize = 15)
plt.show()
```

```python
# Histogram - Cumulative Relative Frequencies - Population
plt.figure(figsize = (12, 8))
plt.hist(pop, bins = 50, density = True, cumulative = True)
plt.title("Cumulative Relative Frequencies - Population", fontsize = 20)
plt.xlabel("Stock Returns 2017 (in %)", fontsize = 15)
plt.ylabel("Cumulative Relative Frequency", fontsize = 15)
plt.show()
```

### Measures of Central Tendency - Geometric Mean 
for compound intrest/growth rate, Geometric Mean works best for Multi-periods setting

```python
import numpy as np
np.set_printoptions(precision=4, suppress= True)
```

```python
Price_2015_2018 = np.array([100, 107, 102, 110])
```

```python
ret = Price_2015_2018[1:] / Price_2015_2018[:-1] - 1
ret
```

```python
mean = ret.mean()
mean
```

```python
100 * (1 + mean)**3
```

```python
geo_mean = (1 + ret).prod()**(1/ret.size) - 1
geo_mean
```

```python
100 * (1 + geo_mean)**3
```

```python
# Geometric Mean
(110 / 100)**(1/ret.size) - 1 
```

### Log Returns
Log returns are additive over time

```python
import numpy as np
np.set_printoptions(precision=4, suppress= True)
```

```python
Price_2015_2018 = np.array([100, 107, 102, 110])
```

```python
ret = Price_2015_2018[1:] / Price_2015_2018[:-1] - 1
ret
```

```python
100 * (ret + 1).prod()
```

```python
log_ret = np.log(Price_2015_2018[1:] / Price_2015_2018[:-1])
log_ret
```

```python
mean_log = log_ret.mean()
mean_log
```

```python
100 * np.exp(mean_log * 3)
```

```python
add = log_ret.sum()
add
```

```python
100 * np.exp(add)
```
